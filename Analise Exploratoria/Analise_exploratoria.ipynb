## Análise Exploratória de uma base de dados de vendas (AdventureWorks )

```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
#pd.options.display.float_format = '{:20,.2f}'.format # Formatando a saída de dados

# Alterando o estilo
# https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html
plt.style.use("dark_background") #ggplot, bmh, grayscale, dark_background, seaborn-whitegrid
```

```python
# Lendo a base de dados
df = pd.read_excel("AdventureWorks.xlsx") # Base de dados: AdventureWorks.xlsx
```

```python
# Visualizando a base de dados (Primeiras linhas)
df.head(3)
```

```python
# Visualizando a base de dados (últimas linhas) 
df.tail(3)
```

```python
# Visualizando uma amostra da base de dados
df.sample(4)
```

```python
# Número de linhas e colunas da base de dados
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Verificando os tipos de dados
df.info()
```

```python
# Qual a receita total
receita_total = df['Valor Venda'].sum()
display(f'Receita Total = {receita_total:,.2f}')
```

```python
# Qual o custo total?
custo_total = df['Custo Unitário'].mul(df['Quantidade']).sum()
display(f'Custo Total = {custo_total:,.2f}')
```

```python
# Criando uma coluna para o custo
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criando a coluna de custo
```

```python
df.head(3)
```

```python
# Qual o custo Total?
round(df["custo"].sum(), 2)
```

```python
# Como temos a receita (coluna 'Valor Venda') e o custo (coluna 'Custo unitário'), 
# Vamos criar uma coluna de Lucro que será Receita - Custo
df["lucro"]  = df["Valor Venda"] - df["custo"] 
```

```python
df.sample(1)
```

```python
# Lucro Total
round(df["lucro"].sum(),2)
```

```python
# Tempo em dias desde a data da venda até a data do envio
# Criando uma coluna com total de dias para enviar o produto
df['Tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.sample(2)
```

```python
df.info()
```

```python
df.sample(2)
```

```python
# Agora, queremos saber a média do tempo de envio para cada Marca, e 
# para isso precisamos transformar a coluna Tempo_envio em númerica

# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.info()
```

```python
df.sample(2)
```

```python
# Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
# Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

 ### Missing Values (Valores ausentes)

```python
# Verificando se temos dados faltantes
df.isnull().sum()
```

#### E, se a gente quiser saber o Lucro por Ano e Por Marca?

```python
# Vamos Agrupar por ano e marca
LucroAnoMarca = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
pd.options.display.float_format = '{:20,.2f}'.format # Formatando a saída de dados
display(LucroAnoMarca)
LucroAnoMarca.dtypes
```

```python
# Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
display(lucro_ano)
lucro_ano.info()
```

```python
# Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico Total de produtos vendidos
plt.figure(dpi = 100)
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
plt.grid(color='y', 
         linestyle='--', 
         linewidth=2.2, 
         axis='both', 
         alpha=0.5);
# grid: https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.grid.html
# linestyle: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
```

```python
plt.figure(dpi = 100)
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
plt.grid(color='c', 
         linestyle='--', 
         linewidth=2.2, 
         axis='both', 
         alpha=0.5);
# grid: https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.grid.html
# linestyle: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
# Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.sample(3)
```
